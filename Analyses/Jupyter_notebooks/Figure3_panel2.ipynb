```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.patches as mpatches
from pywaffle import Waffle

# Read the CSV file into a DataFrame
data = pd.read_csv('../../Data/Fig3/Talos_solves_VCGS-prospective.csv', header=0, index_col=[1, 0], usecols=range(1, 7))

# Combine duplicate rows by summing their values
# This is needed because the data breaks out some additional categories that we are not interested in
data_combined = data.groupby(level=[0, 1]).sum()

# Sorting dataframe so that legends appear in the desired order
custom_sort={'Gene-disease Associations': 1, 'Variant Reclassifications': 2, 'Missed in Manual Analysis': 3, 'CNV/SV': 4}
data_sorted = data_combined.sort_index(key=lambda x: x.map(custom_sort))

colors_list = ['#EC4E20', #flame orginal
        '#F69A7F', #flame light
        '#FCCCBE', #flame lightest
        '#FF9505', #orange original
        '#FFC474', #orange light
        '#016FB9', #azul original
        '#4799D0', #azul light
        '#8EC4E8', #azul lighter
        '#D4EEFF', #azul lightest
        '#57ABA9', #teal light
        '#C2FCFA'] #teal lightest

#print([f"{k} ({v})" for k, v in data['New association'].items()][0:3])

fig = plt.figure(
    FigureClass=Waffle,
    plots={
        411: {
            'values': data_combined.loc[('Gene-disease Associations'),'Renal'],
            'colors': colors_list[0:3],
            #'title': {'label': 'Gene-disease associations', 'loc': 'left', 'fontsize': 12}
        },
        412: {
            'values': data_combined.loc[('Variant Reclassifications'),'Renal'],
            'colors': colors_list[3:5],
            #'title': {'label': 'Variant reclassifications', 'loc': 'left', 'fontsize': 12}
        },
        413: {
            'values': data_combined.loc[('Missed in Manual Analysis'),'Renal'],
            'colors': colors_list[5:9],
            #'title': {'label': 'Missed in original analysis', 'loc': 'left', 'fontsize': 12}
        },
        414: {
            'values': data_combined.loc[('CNV/SV'),'Renal'],
            'colors': colors_list[9:12],
            #'title': {'label': 'CNV/SV', 'loc': 'left', 'fontsize': 12}
        },
    },
    rows=5,
    #colors=colors_list,
    rounding_rule='ceil',
    figsize=(5, 5)
)

import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

patches = []
for x in range(len(data_sorted.index.get_level_values(1))):
    labels = list(data_sorted.index.get_level_values(1))
    patches.append(mpatches.Patch(color=colors_list[x], label=labels[x]))
plt.legend(handles=patches, bbox_to_anchor=(3,2), loc="lower left")

legend1 = plt.legend(handles=patches[0:3], bbox_to_anchor=(1.5,4.3), loc="upper left")
legend2 = plt.legend(handles=patches[3:5], bbox_to_anchor=(1.5,3.0), loc="upper left")
legend3 = plt.legend(handles=patches[5:9], bbox_to_anchor=(1.5,2.1), loc="upper left")
legend4 = plt.legend(handles=patches[9:11], bbox_to_anchor=(1.5,0.8), loc="upper left")
fig.add_artist(legend1)
fig.add_artist(legend2)
fig.add_artist(legend3)
fig.add_artist(legend4)

fig.savefig('../../Figures/Fig3/Fig3_panel2_Renal.png')
```
