```python
from pypdf import PdfReader, PdfWriter

panel_paths = [
    "../../Figures/Fig3/Fig3_panel2.pdf",
    "../../Figures/Fig3/Fig3_panel2_NDD.pdf",
    "../../Figures/Fig3/Fig3_panel2_Cardiac.pdf",
    "../../Figures/Fig3/Fig3_panel2_Renal.pdf",
]
pages = [PdfReader(p).pages[0] for p in panel_paths]

# Calculate original size and add margin (e.g., 15% top/bottom, 10% left/right)
orig_w = sum(p.mediabox.width for p in pages)
orig_h = max(p.mediabox.height for p in pages)
margin_x = orig_w * 0.10
margin_y = orig_h * 0.15
new_w = orig_w + 2 * margin_x
new_h = orig_h + 2 * margin_y

writer = PdfWriter()
page = writer.add_blank_page(width=new_w, height=new_h)

x = margin_x
y = margin_y  # place panels above bottom margin; adjust if you want more top space
for p in pages:
    page.merge_translated_page(p, tx=x, ty=y, expand=False)
    x += p.mediabox.width

combo_path = "../../Figures/Fig3/Fig3_panel2_All.pdf"
with open(combo_path, "wb") as f:
    writer.write(f)


from reportlab.pdfgen import canvas
import tempfile, os
from pypdf import PdfReader, PdfWriter

overlay_file = tempfile.NamedTemporaryFile(delete=False, suffix=".pdf")
c = canvas.Canvas(overlay_file.name, pagesize=(new_w, new_h))
c.setFont("Helvetica-Bold", 12)

def rel_text(x_rel, y_rel, text, bold=False, size=12):
    c.setFont("Helvetica-Bold" if bold else "Helvetica", size)
    c.drawCentredString(x_rel * new_w, y_rel * new_h, text)

def rel_multiline(x_rel, y_rel, text, line_height=14, bold=False, size=12):
    c.setFont("Helvetica-Bold" if bold else "Helvetica", size)
    lines = text.splitlines()
    # draw top-aligned; adjust offset if you want center/other alignment
    for i, line in enumerate(lines):
        c.drawCentredString(x_rel * new_w, (y_rel - i * line_height / new_h) * new_h, line)

# Use relative coords that land in the added whitespace (example positions)
rel_multiline(0.05, 0.92, "New Diagnoses\n(Count)", bold=True)
rel_multiline(0.17,  0.92, "Full Cohort\n(248)")
rel_multiline(0.33, 0.92, "NDD Sub-cohort\n(91)")
rel_multiline(0.53, 0.92, "Cardiac Sub-cohort\n(54)")
rel_multiline(0.725, 0.92, "Renal Sub-cohort\n(43)")
rel_multiline(0.05, 0.79, "Gene-disease\nAssociations")
rel_multiline(0.05, 0.60, "Variant\nReclassifications")
rel_multiline(0.05, 0.41, "Missed in\nManual Analysis")
rel_multiline(0.05, 0.21, "CNV/SV")
rel_multiline(0.05, 0.09, "Total Cohort Size\nNew Diagnosis %", bold=True)
rel_multiline(0.17,  0.09, "N=4738\n5.2%")
rel_multiline(0.33, 0.09, "N=1665\n5.4%")
rel_multiline(0.53, 0.09, "N=997\n5.4%")
rel_multiline(0.725, 0.09, "N=695\n6.1%")

c.save()

# Merge overlay onto the combined page
base = PdfReader(combo_path)
overlay = PdfReader(overlay_file.name)
out = PdfWriter()
pg = base.pages[0]
pg.merge_page(overlay.pages[0])
out.add_page(pg)
with open(combo_path, "wb") as f:
    out.write(f)
os.unlink(overlay_file.name)
```
